**Option 1: Edge Detection** This algorithm attempts to identify edges of objects in an image by finding places where the intensity (darkness/lightness) changes abruptly.  Scan through the pixels in a picture row by row and compare the grayscale value of each pixel with the pixel to its right (and possibly above or below).  If the grayscale value changes by a large amount, then color the pixel white, otherwise color it black. 

Write code for this and test it with an image of your choice. 

```python
image = data.cat()

if len(image.shape) == 2:
    print("Gray image")
else:
    gray = color.rgb2gray(image)

horizontal = np.abs(np.roll(gray, -1, axis=1) - gray)
vertical = np.abs(np.roll(gray, -1, axis=0) - gray)

edge = np.sqrt(horizontal**2 + vertical**2)

edge /= edge.max()

plt.imshow(edge, cmap="gray")
plt.axis("off")
plt.show()
```
