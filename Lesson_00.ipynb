## **Task 1:** Python lists and list comprehensions 

We will work with lists and associated data structures in the class a lot!  

```python
# Recall lists in python ... 
my_list = [3, 5, 1, 2, 4, 8, 9, 7, 6]
print(my_list)
```

```python
my_list[-3]
```

```python
# use a loop to create a list of even numbers from 0 to 20
list_1 = []
for x in range(21):
    if x%2==0:
        list_1.append(x)
        
list_1
```

```python
# we can use the : operator for slicing

print(list_1[:])
print(list_1[2:6])
print(list_1[5:])
print(list_1[:5])
print(list_1[-4:])
```

```python
# we can create new lists compactly using the 'list comprehension' framework
# [x for x in "thing" if "condition"]

new_list=[x for x in range(21) if x%2==0]
new_list
```

----------------
We can also use this for filtering data ... start with a list of names

create new lists from this with
- `names_1` = all names that contain the letter 'e'
- `names_2` = all names that start with 'C'
- `names_3` = all names that end in 'y'
- `names_4` = all names that are 4 letters long
- `names_5` = all names that end in a vowel


```python
names=['Alice','Alvin','Barb','Chuck','Cathy',
       'Dave','Elaine','Fred','Gregory','Henry','Jacob',
       'Ken','Linda','Matty','Ollie','Raul','Sal','Valarie']

names_1 = [n for n in names if 'e' in n]
print(names_1)

names_2 = [n for n in names if n[0]=='C']
print(names_2)

names_3 = [n for n in names if n[-1] =='y']
print(names_3)
names_4 = [x for x in names if len(x) == 4]
print(names_4)

names_5 = [n for n in names if n[-1] in ['a','e','i','o','u']]
print(names_5)
```

```python
# define an array of 1000 random integers in the range 1-100 and use list comprehensions to 
# create lists of the even and odd values from the original.  Are about half of the values even/odd? 

data = np.random.randint(1,100,50)

data_even = [n for n in data if n%2==0]
data_odd = [n for n in data if n%2==1]

print('all = ',data,'\n')
print('evens = ',data_even,'\n')
print('odds = ',data_odd,'\n')
```

```python
# define an array of 1000 standard normal random values and use list comprehensions to 
# create lists of the positive and negative values from the original.  Are about half of the values pos/neg?   

normal_data=np.random.randn(1000)

data_pos = [n for n in normal_data if n >= 0]
data_neg = [n for n in normal_data if n < 0]

#print('All', normal_data)
print("Positive numbers: ", data_pos)
print("Negative numbers: ", data_neg)
```

## **Task 3:** Multidimensional arrays - indexing and slicing

```python
# create an array from the following list ... note how multidimensional indexing is a bit different here

my_array = np.array(my_list)

type(list(my_array))
```

```python
# What do each of the following return?

print(my_array[0,0])
print(my_array[0,:])
print(my_array[2,1:])
print(my_array[:,2])
print(my_array[:3,0])
print(my_array[2:,-1])
```
