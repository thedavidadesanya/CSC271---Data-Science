```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
walk_list = [np.array([0,0])]  #walk list as been initialized to starting position (0,0)

for i in range(100): # run loop 100 times ---> taking 100 steps
    step=np.random.randn(2) #normal distribution that gives two values ---> random 2D steps
    walk_list.append(walk_list[-1]+step) #last walk is taken + step, then appended for the new position

walk_array = np.array(walk_list) #converting walk_list back into an array


x=walk_array[:,0]  #slicing for the first columns ---> x coordinates
y=walk_array[:,1]  #slicing for the second column ----> y coordinates

def distance(walk_array):
    """ This function computes the distance from the origin to each step of the random walk."""
    return np.sqrt(walk_array[:,0]**2 + walk_array[:,1]**2)

def find_steps(walk_array):
    """Finds the length of each step in the random walk"""
    steps = np.diff(walk_array, axis=0)
    return np.sqrt(np.sum(steps**2, axis=1))





my_distance = distance(walk_array)
my_steps = find_steps(walk_array)

median_step = np.median(my_steps)
mean_step = np.mean(my_steps)

#plots the random walk
plt.rcParams["font.family"] = "Times New Roman"
plt.style.use("dark_background")


plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(x,y, color="pink")
plt.scatter(0,0,s=30, color="green", alpha=0.7, linewidths=2)
plt.scatter(x[1:],y[1:],s=20, color="gray")
plt.scatter(x[-1], y[-1], s=30, color="red")
plt.text(0,0,"Origin")
plt.text(x[-1], y[-1], "End")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Random Walk")
plt.grid(True)


plt.subplot(1,3,2)
plt.plot(my_distance)
plt.title("Distance from Origin vs. Step")
plt.xlabel("Step")
plt.ylabel("Distance")
plt.grid(True)


# Plot histogram of step sizes
plt.subplot(1, 3, 3)
plt.hist(my_steps, bins=20, edgecolor='violet', alpha=0.9, color="deepskyblue")
plt.axvline(mean_step, color="red", linestyle="dashed", linewidth=1, label=f"Mean: {mean_step:.3f}")
plt.axvline(median_step, color="lime", linestyle="dashed", linewidth=1, label=f"Median: {median_step:.3f}")
plt.xlabel("Step Size")
plt.ylabel("Rate")
plt.title("Histogram of Step Sizes")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
