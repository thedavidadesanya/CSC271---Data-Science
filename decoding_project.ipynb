```python
def decode(secret):
    """This function decrypts the encrypted message by 
    shifting each letter back by one using ord() and chr() functions by
    subtracting the ascii value by one and returning the new character"""
    return "".join(chr(ord(x)-1) if x!= '\n' else x for x in secret)

#the secret message is a really simple caeser cipher shifitng each letter up by just one place
with open("data/secret_message.txt", "r", encoding="utf-8") as f_in, open("data/secret_message_decoded.txt", "a", encoding="utf-8") as decode_file:
    for line in f_in:
        decode_line = decode(line)
        decode_file.write(decode_line)
        print(line)
        print(decode_line)
```
